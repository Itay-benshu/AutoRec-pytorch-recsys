```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('../data/am_preprocessed.csv')
print(df.shape)
df.head(5)
```

```python
plt.figure(figsize=(18, 4))
plt.hist(df['user_id'].value_counts().values, bins=750)
plt.xlabel('# reviews')
plt.xlim(0, 200)
plt.xlim()
plt.ylabel('# users')
plt.title('Distriution of number of reviews per user')
plt.show()
```

We see a fast decline from the minimum required reviews for the dataset (5) up until ~25

```python
plt.figure(figsize=(18, 4))
plt.hist(df['item_id'].value_counts().values, bins=500)
plt.xlabel('# reviews')
plt.ylabel('# item')
plt.xlim(0, 200)
plt.title('Distriution of number of reviews per item')
plt.show()
```

Similarly, we see a fast decline until ~40.

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='rating')
plt.title('Rating counts')
plt.show()
```

Seems like a strong positive bias towards rated items, as the most popular rating is by far 5 and 4 being the second most.

```python
plt.figure(figsize=(8, 4))
plt.hist(df.groupby('user_id').mean()['rating'], bins=5)
plt.title('Mean user rating distribution')
plt.show()
```

The mean user rating behaves very similarily to the global rating.

```python
plt.figure(figsize=(8, 4))
plt.hist(df.groupby('item_id').mean()['rating'], bins=10)
plt.title('Mean item rating distribution')
plt.show()
```

Again, the mean item rating distribution follows a similar trend.
