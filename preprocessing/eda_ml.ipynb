```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('../data/ml_1m_preprocessed.csv')
print(df.shape)
df.head(5)
```

```python
plt.figure(figsize=(18, 4))
plt.hist(df['user_id'].value_counts().values, bins=600)
plt.xscale('log')
plt.xlabel('# reviews')
plt.xticks(np.logspace(1, 5, 5))
plt.ylabel('# users')
plt.title('Distriution of number of reviews per user')
plt.show()
```

And we see that the vast majority of users have a little more than ~20 reviews (the minimum for the dataset), with a long tail of users with much more reviews.

```python
plt.figure(figsize=(18, 4))
plt.hist(df['item_id'].value_counts().values, bins=2000)
plt.xscale('log')
plt.xlabel('# reviews')
plt.xticks(np.logspace(1, 5, 5))
plt.ylabel('# item')
plt.title('Distriution of number of reviews per item')
plt.show()
```

Similarly we see that the vast majority of items have a little more than 10 reviews, again with a pretty long tail of items with much more reviews.

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='rating')
plt.title('Rating counts')
plt.show()
```

We see the most common ratings are 3 and 4, 5 is also relatively popular, 1 and 2 are rare. Interestingly we also see non-integer reviews (0.5, 1.5 etc.), but they are much less common.

```python
plt.figure(figsize=(8, 4))
plt.hist(df.groupby('user_id').mean()['rating'], bins=20)
plt.title('Mean user rating distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df.groupby('item_id').mean()['rating'], bins=20)
plt.title('Mean item rating distribution')
plt.show()
```

Not surprisingly we see that most users / items have a mean rating around 3-4. However especially for the items we see a significant amount of items which are broadly rated lower, indicating low-quality / unpopular movies.
